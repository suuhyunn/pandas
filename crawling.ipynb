```python
!pip install selenium

from selenium import webdriver      # 웹 연결
from bs4 import BeautifulSoup as bs         # 분석하기 쉽게 정리
import pandas as pd         # 분석 도구 라이브러리
```

```python
driver = webdriver.Chrome('chromedriver.exe')
url = "https://www.starbucks.co.kr/store/store_map.do"
driver.get(url)
```

```python
area_btn = '#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a'
driver.find_element_by_css_selector(area_btn).click()
```

```python
seoul_btn = '#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a'

driver.find_element_by_css_selector(seoul_btn).click()
```

```python
all_btn = '#mCSB_2_container > ul > li:nth-child(1) > a'

driver.find_element_by_css_selector(all_btn).click()
```

```python
html = driver.page_source
soup = bs(html,'html.parser')
soup
```

```python
starbucks_list = soup.select('li.quickResultLstCon')
starbucks_list
```

```python
starbucks_list[0]
```

```python
star = starbucks_list[0]

name = star['data-name']    # 딕셔너리 형태는 value값 뽑아내기 쉬움
lat = star['data-lat']
lng = star['data-long']
addr = str(star.select('p')[0]).split('<br/>')[0].split('>')[1]
phone = str(star.select('p')[0]).split('<br/>')[1].split('<')[0]
phone
```

```python
star_list = []

for item in starbucks_list:
    name = item['data-name']
    lat = item['data-lat']
    lng = item['data-long']
    addr = str(item.select('p')[0]).split('<br/>')[0].split('>')[1]
    phone = str(item.select('p')[0]).split('<br/>')[1].split('<')[0]
    star_list.append([name, lat, lng, addr, phone])
```

```python
star_list
```

```python
starbucks_df = pd.DataFrame(star_list, columns=['매장명', '위도', '경도', '주소', '전화번호'])
starbucks_df.head()
```

```python
starbucks_df.to_excel('starbucks_list.xlsx', index=False)   #엑셀 저장할때는 늘 인덱스 날리고 저장해야 함
```

```python
sgg_names = []

for item in starbucks_df['주소']:
    sgg = item.split()[1]
    sgg_names.append(sgg)
```

```python
starbucks_df['시군구명'] = sgg_names
starbucks_df.head()
```

```python
starbucks_count = starbucks_df.pivot_table(index='시군구명',
                                           values='매장명',
                                           aggfunc='count')
                                           
```

```python
# 기본 지도 그리기
import folium
starmap = folium.Map(
    location=[37.5665, 126.9780], 
    zoom_start=11,
    tiles = 'Stamen Terrain')
starmap
```

```python
lat = 37.5665
lng = 126.9780

folium.Marker(
    location = [lat, lng]
).add_to(starmap)

starmap
```

```python
for idx in starbucks_df.index:
    lat = starbucks_df.loc[idx,'위도']
    lng = starbucks_df.loc[idx,'경도']

    folium.Marker(
        location = [lat, lng]
    ).add_to(starmap)

starmap
```
